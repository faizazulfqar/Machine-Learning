```python
#Reading dataset
homelessness=pd.read_csv('homelessness.csv',index_col=0)
print(homelessness)
```

```python
#Sort homelessness by the number of by number of homeless individuals, from smallest to largest and save this as homelessness_ind
homelessness_ind=(homelessness.sort_values("individuals"))
homelessness_ind.head()
```

```python
#Sort homelessness by the number of by number of homeless individuals, in descending order and save this as homelessness_ind
homelessness_fam=(homelessness.sort_values("individuals",ascending=False))
homelessness_fam.head()
```

```python
#sort homelessness first by region(ascending) and then by family numbers(descending)
homelesssness_reg_fam=homelessness.sort_values(["region","family_members"],ascending=[True,False])
homelesssness_reg_fam.head(10)
```

```python
#Subsetting
individuals=homelessness["individuals"]
print(individuals.head())
print("_____"*15)
individuals=homelessness[["individuals","region"]]
print(individuals.head())
```

```python
#Filter homelessness where the number of individuals are greater than 10000
ind_get=homelessness[homelessness["individuals"]>10000]
ind_get.reset_index(0)
```

```python
#Filter homelessness where USA region is Mountain
mountain_reg=homelessness[homelessness["region"]=="Mountain"]
mountain_reg.reset_index(0)
```

```python
#Filter where USA is "South Atlantic" or "Mid-Atlantic"
south_mid_atlantic=homelessness[(homelessness["region"]=="South Atlantic")|(homelessness["region"]=="Mid-Atlantic")]
south_mid_atlantic
```

```python
#Filter for rows where family_members is less than 1000 and region is "Pacific"
Ind_Pacific=homelessness[(homelessness["family_members"]<1000) & (homelessness["region"]=="Pacific")]
Ind_Pacific
```

```python
#Filter for cases where USA state is in the list of Mojave states, canu
#The Mojave Desert states are
canu=["California","Arizona","Nevada","Utah"]
#Filter for rows in Mojave state
mojave_homeless=homelessness[(homelessness["state"].isin(canu))]
mojave_homeless
```

```python
#Aggregating another dataset
Temperature=pd.read_csv("temperatures.csv")
Temperature.head()
```

```python
#Sort the index of temperature_ind
temperature_ind=Temperature.set_index(["country","city"])
temp_sort=temperature_ind.sort_index()
temp_sort
```

```python
#Slicing to get subset from Pakistan to Russia with .loc[]
temp_sort.loc["Pakistan":"Russia"]
```

```python
#subset from Lahore to Moscow
temp_sort.loc["Lahore":"Moscow"]
```

```python
#Subset from Pakistan,Lahore to Russia,Moscow
temp_sort.loc[("Pakistan","Lahore"):("Russia","Moscow")]
```

```python
#Slicing for 1st 5 rows
Temperature.iloc[:5,:]
```
