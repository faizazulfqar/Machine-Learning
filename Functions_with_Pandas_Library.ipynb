##Importing Libraries

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Numpy Library is used to create arrays and matrices. It makes homogeneous objects.
Pandas Library is used for data manipulation and data analysis. It is used for creating hetrogeneous objects.
Matplotlib is a plotting library used for visualization of data. Seaborn is also a data visualization library used for plotting based on matplotlib. And it provides highr interface than matplotlib. Seaborn also hold various default datasets.


##Loading a built-in dataset

```python
#To check which datasets are available
DataFrame=sns.get_dataset_names()
DataFrame
```

```python
#To load a built_in dataset
Flight_DF=sns.load_dataset('flights')
Flight_DF
```

##Some Basic Functions

```python
#To Access first 5 rows of dataset
Flight_DF.head(5)
```

Head function is used to access first rows from dataset. Any number of rows can be accessed.

```python
#To access last 5 rows of dataset
Flight_DF.tail(5)
```

Tail function is used to show the last rows of dataset. Any number of rows can be accessed.

```python
#To check shape of data
Flight_DF.shape
```

144 is number of rows in dataset and 3 is number of columns in dataset.

```python
#To access a single coulum/series from dataset
Flight_DF['passengers']
```

```python
#To access two columns
Flight_DF[['passengers','month']]
```

```python
#To access a specific row
Flight_DF.loc[4]
```

```python
#To access more than one row with loc function
Flight_DF.loc[4:8]
```

Loc function is used for label indexing. It includes start and end position.

```python
#To access more than one row with iloc function
Flight_DF.iloc[4:8]
```

iloc function is used to access rows based on integer location indexing. It doesn't include last position.

```python
#To access specific rows and columns
Specific_Data=Flight_DF.iloc[0:5,:]
Specific_Data
```

The above code prints first 5 rows and all columns.

```python
#To access 3rd row
Flight_DF.iloc[2]
```

```python
#To check the summary of dataset
Flight_DF.info()
```

```python
#To Find Descriptive Statistics of dataset
Flight_DF.describe()
```

```python
#Filtering using a boolean condition
Filtered_Data=Flight_DF[Flight_DF['passengers']>500]
print(Filtered_Data)
```

```python
#To count unique values in a single column
counts=Flight_DF['passengers'].value_counts()
counts
```

```python
#Data visualization
Flight_DF['passengers'].plot(kind='line')
```

```python
#Sorting DataFrame by a column
Flight_DF.sort_values(by='passengers')
```

```python
#Creating a DataFrame
New_Data={
    'Name':["Peter","John","Anna","Joseph","Lisa","Bella"],
    'Age':[28,30,26,24,29,33],
    'City':["Florida","New York","California","Boston","Dallas","Austin"]
}
#Coverting to DataFrame
df=pd.DataFrame(New_Data)
print(df)
```

```python
#Adding a new column
df['Degree']=["MSCS","MBA","BS Math","BBA","BSSE","MIT"]
df
```

```python
#To access starting 2 rows
print(df)
print("_________________________________")
print(df.head(2))
```

```python
#Dropping a column
df.drop("Degree",axis=1)
```

```python
#Creating another DataFrame
Data2={
 'City':["Florida","New York","California",None,"Dallas","Austin"],
 'Population':[87043,102347,47892,18530,None,54901]
}
df2=pd.DataFrame(Data2)
df2
```

```python
#Merging 2 DataFrames based on one common column
merged_df=pd.merge(df,df2,on='City')
merged_df
```

All the rows with same data will be printed. A row that has 'None' value will be skippes. Merging will be based on only commen data.

```python
#Handling Missing Values of df2
print(df2)
print("___________________________________________________")
df2_filled=df2.fillna({'City':'Boston','Population':'73109'})
print(df2_filled)
```

```python
#Dropping row with missing value
print(df2)
print("___________________________")
df2_dropped=df2.dropna()
print(df2_dropped)
```
